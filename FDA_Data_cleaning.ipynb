```python
#1.Insert the symbol '-' into the values in the column and change the datatype.

#Create an empty list
recall_initiation_datelist = []

#Iterate the column and input the symbol '-'
for i in df['recall_initiation_date']:
    recall_init_date = i[0:4] + '-' + i[4:6] + '-' + i[6:8]
    recall_initiation_datelist.append(recall_init_date)
recall_initiation_datelist
```

```python
#Create a new column using the list
df['recall_initiation_date2'] =  recall_initiation_datelist
```

```python
#Create an empty list
center_classification_datelist = []

#Iterate the column and input the symbol '-' 
for j in df['center_classification_date']:
    if not pd.isna(j):
        cent_class_date = str(j)[0:4] + '-' + str(j)[4:6] + '-' + str(j)[6:8]
        center_classification_datelist.append(cent_class_date)
    else:
        center_classification_datelist.append(None)
center_classification_datelist
```

```python
#Create a new column using the list
df['center_classification_date2'] =  center_classification_datelist
```

```python
#Create an empty list
report_datelist = []

#Iterate the column and input the symbol '-'
for k in df['recall_initiation_date']:
    report_date = k[0:4] + '-' + k[4:6] + '-' + k[6:8]
    report_datelist.append(report_date)
report_datelist
```

```python
#Create a new column using the list
df['report_date2'] =  report_datelist
```

```python
#Create an empty list
product_measurement = []
for i in df1['product_measurement']:
    if 'unit' in str(i):
        ans = 'units'
    elif 'case' in str(i):
        ans = 'cases'
    elif 'piece' in str(i):
        ans = 'pieces'
    elif 'bottle' in str(i):
        ans = 'bottles'
    elif 'Ibs' in str(i):
        ans = 'Ibs'
    elif 'case' in str(i):
        ans = 'cases'   
    elif 'can' in str(i):
        ans = 'cans'
    elif 'bar' in str(i):
        ans = 'bars'
    elif 'quantity' in str(i):
        ans = 'quantity'
    elif 'gallon' in str(i):
        ans = 'gallons'
    elif 'carton' in str(i):
        ans = 'cartons'
    elif 'jar' in str(i):
        ans = 'jars'
    elif 'containers' in str(i):
        ans = 'containers'    
    elif 'pound' in str(i):
        ans = 'pound'
    else:
        ans = i
    product_measurement.append(ans)
```

```python
#Create a new column having the product measurement
df1['Product_measurement'] = product_measurement
```
